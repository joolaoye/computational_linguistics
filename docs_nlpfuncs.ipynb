Importing Dependencies

```python
import librosa as lb
import matplotlib.pyplot as plt
```

Librosa is widely used for audio processing, in this notebook I will be using it to extract the digital representation
of the audio waveform and the sampling rate

Matplotlib helps with data visualization

Loading the audio file 

```python
audio_data, sr = lb.load("Sample.wav",sr=None)  
```

Converting the digital data extracted from a WAV file into a magnitude spectrogram and visualizing it

```python
from nlpfuncs import convert_to_spectrogram
mag = convert_to_spectrogram(audio_data)
lb.display.specshow(mag)
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

Converting a magnitude spectrogram into a Mel spectrogram and visualizing it

```python
from nlpfuncs import convert_to_mel
mel = convert_to_mel(mag,sr)
lb.display.specshow(mel)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.show()
```

Converting the digital data extracted from a WAV file into Mel-frequency cepstral coefficients (MFCC) and visualizing it

```python
from nlpfuncs import convert_to_mfcc
mfcc = convert_to_mfcc(audio_data,sr)
lb.display.specshow(mfcc, sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-frequency cepstral coefficients (MFCC)')
plt.xlabel('Time (seconds)')
plt.ylabel('Frequency (Hz)')
plt.show()

```
